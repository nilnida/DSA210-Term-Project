# Data Visualization

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## Average Charging Time across Cycle Bins

```python
df_merged_charge_and_RUL = pd.read_csv("/content/drive/MyDrive/dsa210/df_merged_charge_and_RUL.csv")

df_merged_charge_and_RUL['Cycle bin'] = pd.cut(df_merged_charge_and_RUL['Cycle index'], bins=np.arange(0, 650, 25))
df_avg_time = df_merged_charge_and_RUL.groupby('Cycle bin')[['Time (secs)']].mean().reset_index()
df_avg_time['Cycle bin label'] = df_avg_time['Cycle bin'].astype(str)

# Barplot
plt.figure(figsize=(8, 5))
sns.barplot(data=df_avg_time, x=df_avg_time['Cycle bin label'], y='Time (secs)', palette='pastel')
plt.plot(df_avg_time['Cycle bin label'], df_avg_time['Time (secs)'], color='grey', linestyle='-', label='Trend')
plt.title('Average Charging Time across Cycle Bins')
plt.xlabel('Cycle Bin')
plt.ylabel('Average Charging Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Point plot
plt.figure(figsize=(8, 5))
sns.pointplot(data=df_avg_time, x='Cycle bin label', y='Time (secs)', color='lightgreen', markers='o', linestyles='-')
plt.title('Average Charging Time across Cycle Bins')
plt.xlabel('Cycle Bin')
plt.ylabel('Average Charging Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Average Battery Capacity across Temperature Bins

```python
df_merged_discharge_and_RUL = pd.read_csv("/content/drive/MyDrive/dsa210/df_merged_discharge_and_RUL.csv")

df_merged_discharge_and_RUL['Temperature bin'] = pd.cut(df_merged_discharge_and_RUL['Battery temperature (C)'], bins=np.arange(31, 34, 0.25))
df_avg_temp = df_merged_discharge_and_RUL.groupby('Temperature bin')[['Battery capacity (Ahr) for discharge till 2.7V']].mean().reset_index()

# Barplot
plt.figure(figsize=(8, 5))
sns.barplot(data=df_avg_temp, x=df_avg_temp['Temperature bin'].astype(str), y='Battery capacity (Ahr) for discharge till 2.7V', palette='pastel')
plt.plot(df_avg_temp['Temperature bin'].astype(str), df_avg_temp['Battery capacity (Ahr) for discharge till 2.7V'], color='grey', linestyle='-', label='Trend')
plt.title('Average Capacity across Temperature Bins')
plt.xlabel('Temperature Bin')
plt.ylabel('Average Capacity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Point plot
plt.figure(figsize=(8, 5))
sns.pointplot(data=df_avg_temp, x='Temperature bin', y='Battery capacity (Ahr) for discharge till 2.7V', color='peachpuff', markers='o', linestyles='-')
plt.title('Average Capacity Across Temperature Bins (Point Plot)')
plt.xlabel('Temperature Bin')
plt.ylabel('Average Capacity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Corralation Anaysis on Discharge Resullts

```python
df_merged_discharge_and_RUL = pd.read_csv("/content/drive/MyDrive/dsa210/df_merged_discharge_and_RUL.csv")

plt.figure(figsize=(8, 5))
corr_matrix = df_merged_discharge_and_RUL[["Battery temperature (C)", "Battery capacity (Ahr) for discharge till 2.7V", "RUL"]].corr()

sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
